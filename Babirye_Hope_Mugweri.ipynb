```python
#Reading the data
data = pd.read_csv('titanic.csv')

#Viewing the first 5 rows of the dataset
data.head()
```

```python
#Extracting values from dataframe apart from the last column 
x = data.iloc[:,:-1].values
```

```python
#running X, dataframe without the survived column
x = data.drop('Survived',axis=1)
x
```

```python
#running the survived column, the independent variable
y = data['Survived']
y
```

```python
#checking presence of missing values in data
x.isnull().sum()
```

```python
#dropping columns having missing values greater than 50% and are not so efficient to predict the person’s survival rate
x.drop(['Name','Ticket','Cabin'],axis=1,inplace=True)
```

```python
#handling missing values

#numeric valueimputation with mean
x['Age'] = x['Age'].fillna(x['Age'].mean())

#categorical value imputation with mode, most frequent category
x['Embarked'] = x['Embarked'].fillna(x['Embarked'].mode()[0])
```

```python
print(x.columns)
```

```python
x = pd.get_dummies(x, columns = ['Sex', 'Embarked'],
                   prefix = ['Sex', 'Embarked'], 
                   drop_first = True)
```

```python
# train-test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, 
                                                    test_size = 0.2,
                                                    random_state = 0)
```
